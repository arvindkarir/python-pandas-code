```python
from pypostalcode import PostalCodeDatabase
pcdb = PostalCodeDatabase()
# https://github.com/inkjet/pypostalcode
```

```python
def get_pc(val):
    try:
        location = pcdb[val]
        return location
    except (AttributeError, IndexError) as e:
        print(val)
```

```python
a_dict = {}
d = []
for index,row in df3.iterrows():
    info = get_pc(row['code'])
    if not info:
        continue
    print(
    info.postalcode,
    info.city,
    info.longitude,
    info.latitude,
    info.timezone)
    a_dict[(info.longitude,info.latitude)] = row['visits']
    d.append({'lon': info.longitude, 'lat' : info.latitude, 'visit': row['visits']})
```

```python
# plotting as pointers
gta_latlong = (43.598311, -79.662868)            
m = folium.Map(location=gta_latlong, zoom_start=12)
locationlist = df4[['lat','lon']].values.tolist()
labels = df4['visit'].values.tolist()
for point in range(len(locationlist)):
    folium.Marker( locationlist[point] ).add_to(m)
m.save( 'map.html')
display(m)
```

```python
# plotting data as marker circles, radius of circle can work as heat map 'ish
gta_latlong = (43.598311, -79.662868)            
m = folium.Map(location=gta_latlong, zoom_start=12,
              control_scale = True,  attr='USGS style')
for index, row in df4.iterrows():
    folium.CircleMarker([row['lat'], row['lon']],
                        radius=25).add_to(m)
display(m)
```

```python
# extracting the data as Series, may not be required
lats = pd.Series(df4['lat']).values
longs = pd.Series(df4['lon']).values
visits = pd.Series(df4['visit']).values
```

```python
# making sure of data type
visits.dtype
```

```python
# converting the data into a list for plotting
my_data = list(zip(lats,longs,visits))
my_data
```

```python
from folium.plugins import HeatMap
# Using USGS style tile
url_base = 'http://server.arcgisonline.com/ArcGIS/rest/services/'
service = 'NatGeo_World_Map/MapServer/tile/{z}/{y}/{x}'
tileset = url_base + service
gta_latlong = (43.598311, -79.662868)            
heatmap_map = folium.Map(location=gta_latlong, zoom_start=12, control_scale = True, tiles=tileset, attr='USGS style')
# folium.Marker( location = [43.7254, 79.4521]).add_to(heatmap_map) # this does not work adding a marker for the Mall
hm = plugins.HeatMap(my_data, radius=20, blur=15, max_val = 1.5, gradient={.4: 'blue', .5: 'lime', 0.65: 'yellow', 0.95: 'orange', 1: 'red'})
m = heatmap_map.add_child(hm)
m.save( 'heat_map.html')
m
```
