```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# datetime parser
# dateparse = lambda x: pd.datetime.strptime(x, '%y/%m/%d %H:%M')
df = pd.read_csv('./Toronto.csv')
```

```python
sns.set(style="darkgrid")
# set size of figure
plt.figure(figsize=(200,30))
# use horizontal stripplot with x marker size of x
sns.stripplot(y='Rating',x='Serial', data=df,
 orient='h', marker='o', color='navy', size=25)
# rotate x tick labels
plt.xticks(rotation=15)
# remover borders of plot
plt.tight_layout()
plt.savefig('fig1.png')
plt.show()
```

```python
df2 = pd.read_csv('./toronto_snow.csv')
```

```python
sns.set(style="darkgrid")
# set size of figure
plt.figure(figsize=(200,30))
# use horizontal stripplot with x marker size of x
sns.stripplot(y='Snow_day',x='Number', data=df2,
 orient='h', marker='>', color='navy', size=15)
# rotate x tick labels
plt.xticks(rotation=15)
# remover borders of plot
plt.tight_layout()
plt.savefig('snow.png')
plt.show()
```
