```python
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
import glob
from skimage.io import imread
```

```python
from math import sqrt
```

```python
image_file = glob.glob('milky_way.jpg')[0]
```

```python
im = imread(image_file, as_grey=True)
plt.imshow(im, cmap='gray')
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.1)
# compute radii in the third column
blobs_log[:, 2] = blobs_log[:, 2]*sqrt(2)
numrows = len(blobs_log)
```

```python
print('Number of stars identified:', numrows)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(im, cmap='gray')
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+1, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)
plt.show()
```
